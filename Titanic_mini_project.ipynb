Hereâ€™s a brief summary of the 14 attributes from titanic dataset:

    pclass: Passenger class (1 = 1st; 2 = 2nd; 3 = 3rd)
    survival: A Boolean indicating whether the passenger survived or not (0 = No; 1 = Yes); this is our target
    name: A field rich in information as it contains title and family names
    sex: male/female
    age: Age, asignificant portion of values aremissing
    sibsp: Number of siblings/spouses aboard
    parch: Number of parents/children aboard
    ticket: Ticket number.
    fare: Passenger fare (British Pound).
    cabin: Doesthe location of the cabin influence chances of survival?
    embarked: Port of embarkation (C = Cherbourg; Q = Queenstown; S = Southampton)
    boat: Lifeboat, many missing values
    body: Body Identification Number
    home.dest: Home/destination

```python
# gantilah nama penumpang yang kosong dengan anonymous
```

```python
titanic_df['name'].fillna('anonymous', inplace=True)
titanic_df
```

```python
# buatlah kolom baru berisi nama keluarga (surname) dari data penumpang titanic.
# hint: nama pada data merupakan gabungan dari [nama keluarga, nama utama] yang dipisah oleh koma. nama pribadi diawali dengan inisial Mrs., Mr. dst.
```

```python
titanic_df['surname'] = titanic_df['name'].str.split(',').str.get(0)
titanic_df
```

```python
# buatlah kolom baru berisi nama pribadi (first name) dari data penumpang titanic.
# hint: nama pada data merupakan gabungan dari [nama keluarga, nama pribadi] yang dipisah oleh koma. nama pribadi diawali dengan inisial Mrs., Mr. dst.
```

```python
titanic_df['first_name'] = titanic_df['name'].str.split(', ').str.get(1)
titanic_df
```

```python
# buatlah kolom baru yang bernama `age_category` dengan syarat:
#1. kanak dengan usia 0-13
#2. remaja 14-40
#3. dewasa 40-60
#4. lansia > 60
#dimana didalam kolom tersebut berisi kategori string [kanak,remaja,dewasa,lansia]
```

```python
usia = [0, 13, 40, 60, 100]
labels = ['kanak', 'remaja', 'dewasa', 'lansia']
titanic_df['age_category'] = pd.cut(titanic_df['age'], bins=usia, labels=labels)
titanic_df
```

```python
penumpang_mr = titanic_df[titanic_df['name'].str.contains('Mr\.', case=False) & ~titanic_df['name'].str.contains('Mrs\.', case=False)]
penumpang_mr
```

```python
import matplotlib.pyplot as plt
```

Proporsi gender penumpang titanic

```python
gender = titanic_df['sex'].value_counts()
plt.pie(gender, labels=gender.index, autopct='%1.1f%%', colors=['blue', 'pink'])
plt.axis('equal')
plt.title('Proporsi Gender Penumpang Titanic')
plt.show()
```

visualisasi data diatas adalah berbentuk pie chart yang menunjukkan proporsi gender penumpang Titanic. Diagram pie chart menunjukkan dengan jelas bahwa ada perbedaan yang cukup signifikan antara jumlah penumpang laki-laki (biru ) dan perempuan (pink). 64,4 % penumpang adalah laki-laki, 35,6% sisanya adalah perempuan.

Jumlah Penumpang Berdasarkan Pelabuhan pemberangkatan

```python
embarked_counts = titanic_df['embarked'].value_counts()
plt.bar(embarked_counts.index, embarked_counts.values, color='lightblue')
plt.xlabel('Pelabuhan')
plt.ylabel('Jumlah Penumpang')
plt.title('Jumlah Penumpang Berdasarkan Keberangkatan Pelabuhan')
plt.show()
```

Diagram batang diatas adalah bar chart yang menampilkan jumlah penumpang berdasarkan port of embarkation (tempat keberangkatan/pelabuhan) di kapal Titanic.kita dapat melihat jumlah penumpang yang berangkat dari masing-masing pelabuhan. Mayoritas penumpang berangkat dari Southampton (S), diikuti oleh Cherbourg (C) dan Queenstown (Q).

Rata- rata tarif berdasrkan kelas penumpang

```python
fare_by_class = titanic_df.groupby('pclass')['fare'].mean()
fare_by_class.plot(kind='bar', color='gray')
plt.xlabel('Kelas Penumpang')
plt.ylabel('Rata-rata Fare')
plt.title('Rata-rata Fare Berdasarkan Kelas Penumpang')
plt.xticks(rotation=0)
plt.show()
```

Diagram batang (bar chart) yang menampilkan rata-rata harga tiket berdasarkan kelas penumpang (Pclass) di kapal Titanic.Dari visualisasi ini, kita dapat melihat rata-rata harga tiket berdasarkan kelas penumpang di kapal Titanic. Kelas 1 memiliki rata-rata harga tiket tertinggi, diikuti oleh kelas 2 dan kelas 3.

Jumlah penumpang yang selamat dan tidak selamat

```python
survival_counts = titanic_df['survived'].value_counts()
plt.figure(figsize=(6, 6))
plt.pie(survival_counts, labels=['Tidak selamat', 'Selamat'], autopct='%1.1f%%', colors=['lightcoral', 'lightgreen'])
plt.axis('equal')
plt.title('Jumlah Penumpang yang Selamat dan Tidak Selamat')
plt.show()
```

Diagram pie chart yang menggambarkan jumlah penumpang yang selamat dan yang tidak selamat di kapal Titanic.kita dapat melihat proporsi jumlah penumpang yang selamat (warna hijau) dan yang tidak selamat (warna merah) di kapal Titanic. Terdapat 61,8% tidak selamat dan 38,2% selamat dalam kejadian tenggelamnya kapal Titanic.

Rata-rata jumlah saudara/pasangan di kapal

```python
rata_rata = titanic_df.groupby('pclass')['sibsp'].mean()
plt.figure(figsize=(8, 6))
rata_rata.plot(kind='bar', color='orange')
plt.xlabel('Kelas Penumpang')
plt.ylabel('Jumlah Rata-rata SibSp')
plt.title('Jumlah Rata-rata SibSp Berdasarkan Kelas Penumpang')
plt.xticks(rotation=0)
plt.show()
```

Diagram batang yang menggambarkan jumlah rata-rata saudara/pasangan (SibSp) yang berada di kapal Titanic untuk setiap kelas penumpang.Dari diagram batang ini, kita dapat melihat bahwa kelas penumpang (Pclass) 2 memiliki rata-rata jumlah saudara/suami-istri (SibSp) yang lebih rendah dibandingkan dengan kelas penumpang 1 dan 3.
